```python
"""
Linear elasticity problems. The example come from OGS benchmark.
"""

from fenics import *

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

$$ \lambda = \frac {E \nu}{(1 + \nu)(1 - 2 \nu)} $$
$$ \mu = \frac {E}{2(1 + \nu)} $$

```python
mod = 1
miu = 0.3
mu = mod / (2 + 2*miu)
lambda_ = (mod*miu)/((1 + miu)*(1 - 2*miu))
```

### creat geometry and mesh

```python
mesh = UnitSquareMesh(8, 8)
plot(mesh)
```

```python
V = VectorFunctionSpace(mesh, 'P', 1)
```

### Define boundaries

```python
left= 'near(x[0], 0)'
right = 'near(x[0], 1)'
top = 'near(x[1], 1)'
bottom = 'near(x[1], 0)'
```

### Set buundary condition

```python
bc_top = DirichletBC(V, Constant((0, 0.05)), top)
bc_left = DirichletBC(V, Constant((0, 0)), left)
bc_bottom = DirichletBC(V, Constant((0, 0)), bottom)
bc = [bc_top, bc_left, bc_bottom]
```

### Define Strain and Stress

#### Strain and displacement relationship

$$ \varepsilon_i,_j = \frac{1}{2}  (u_i,_j + u_j,_i) $$

#### Constitutive modules
    
$$ \sigma_i,_j = 2\nu\varepsilon_i,_j  + \lambda tr(\varepsilon_i,_j)\mathbf I $$

```python
def epsilon(u):
    return 0.5*(nabla_grad(u) + nabla_grad(u).T)

def sigma(u):
    return lambda_* tr(nabla_grad(u))*Identity(d) + 2*mu*epsilon(u)
```

### Define variational problems
#### Equilibrium equation
$$ \nabla\sigma + f = 0 $$

#### Trial function 
$$ - \int_\Omega \nabla\sigma \cdot v d \Omega= \int_\Omega f \cdot v d\Omega  $$

#### In Fenics

$$ - \int_\Omega \nabla\sigma \cdot v dx= \int_\Omega f \cdot v dx  $$

#### Intergration by parts or the theory of virtual work

$$ - \int_\Omega \nabla\sigma \cdot v dx = \int_\Omega \sigma : \nabla v dx - \int_{\partial_\Omega} (\sigma \cdot n)\cdot v ds  $$

#### we could get the weak form 
$$ \int_\Omega \sigma : \nabla v dx = \int_\Omega f \cdot v dx + \int_{\partial_\Omega} (\sigma \cdot n)\cdot v ds $$

$$ \int_\Omega \sigma : \varepsilon dx = \int_\Omega f \cdot v dx + \int_{\partial_\Omega} (\sigma \cdot n)\cdot v ds $$

#### $T = \sigma \cdot n$ is the stress vector at the boundary

####  Got the variational form

$$ a(u, v) = L(v) $$

#### $ \sigma $ and $\varepsilon $ related to $ u $ and $ v $

```python
u = TrialFunction(V)
d = u.geometric_dimension()  # space dimension
v = TestFunction(V)
```

### Ignore gravity and no stress boundary condition

```python
f = Constant((0, 0))
T = Constant((0, 0))
```

```python
a = inner(sigma(u), epsilon(v))*dx
L = dot(f, v)*dx + dot(T, v)*ds
```

### Compute solution

```python
u = Function(V)
solve(a == L, u, bc)
```

```python
# Plot solution
plot(u, title='Displacement', mode='displacement')
```

#### The equation of deviatoric stress
$$ s = \sigma -\frac{1}{3}tr(\sigma)\mathbf I $$

#### The equation of Von_Mises stress

$$ \sigma_M = \sqrt{\frac{3}{2} s : s} $$

```python
# Plot stress
s = sigma(u) - (1./3)*tr(sigma(u))*Identity(d)  # deviatoric stress
von_Mises = sqrt(3./2*inner(s, s))

von_Mises = project(von_Mises, V)
plot(von_Mises, title='Stress intensity')
```

```python
# Compute magnitude of displacement
u_magnitude = sqrt(dot(u, u))
u_magnitude = project(u_magnitude, V)
plot(u_magnitude, 'Displacement magnitude')
```

```python
vertex_values_u_magnitude = u.compute_vertex_values(mesh)
print('min/max u:',
      vertex_values_u_magnitude.min(),
      vertex_values_u_magnitude.max())

# Save solution to file in VTK format
File('elasticity/displacement.pvd') << u
File('elasticity/von_mises.pvd') << von_Mises
File('elasticity/magnitude.pvd') << u_magnitude

# Hold plot
plt.show()
plt.interactive(True)
```
