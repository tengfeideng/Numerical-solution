### Equation of motion of the basic dynamic system


$$ m\ddot u(t) + c \dot u(t) + k u(t) = p(t)  $$

we assume:
$$ u_1 = u $$


$$  u_2 = \dot u(t) $$

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
# Input constants 

delta_t = 0.02 # time step size (seconds)
t_max = 20 # max sim time (seconds)
theta1_0 = np.pi/2 # initial angle (radians)
theta2_0 = 0 # initial angular velocity (rad/s)
theta_init = (theta1_0, theta2_0)
# Get timesteps
t = np.linspace(0, t_max, t_max/delta_t)
```

```python
def int_pendulum_sim(theta_init, t, L=2, m=1, c=0.1, k=9.81):
    theta_dot_1 = theta_init[1]
    theta_dot_2 = -c/m*theta_init[1] - k/m* theta_init[0] + 1/m * np.sin(t)
    return theta_dot_1, theta_dot_2
```

```python
theta_vals_int = integrate.odeint(int_pendulum_sim, theta_init, t)
plt.plot(theta_vals_int)
plt.grid()
```

```python
g = 9.8
l = 1
def fun2(d_list, t):
    omega, theta = d_list
    return np.array([-g/l*theta, omega])
t = np.linspace(0, 20, 2000)


result = integrate.odeint(fun2, [0, 35/180*np.pi], t)

plt.plot(t, result[:, 0])
plt.plot(t, result[:, 1])
plt.grid()
plt.show()
```
